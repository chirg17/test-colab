Question 1 - Data Analysis

```python
# Creating a pivotable displaying the average tip amount spent each day for each gender

df_tips.pivot_table(values='tip', index='day', columns='sex', aggfunc=['mean','std'])
```

Notes: This study examines how men's and women's tipping behaviours alter over time, and if these differences reflect gender or economic position

Question 2 - Advanced Data Filtering

```python
# Using loc and Boolean indexing filtering the data for women who did not survive the crash and were in the third class.

df_ttnc.loc[(df_ttnc['Sex'] == 'female') & (df_ttnc['Survived'] == 0) & (df_ttnc['Pclass'] == 3),["PassengerId","Name","Age","Fare"]]
```

Question 3 - Data Transformation and Aggregation

```python
# Creating a new columns named as "Total Bill per Person”

df_tips['Total Bill per Person'] = df_tips['total_bill']/df_tips['size']
```

```python
# Creating another columns named as "Tip Ratio," which represents the proportion of the tip amount to the total bill.

df_tips['Tip ratio'] = df_tips['tip'] /df_tips['total_bill']
```

```python
df_tips
```

Note: The two columns, "Total Bill per Person" and "Tip Ratio," provide a more detailed view of spending habits by allowing for individual-level analysis inside groups and insight into tipping behavior about the total cost.The "Total Bill per Person" column sheds information on individual contributions by highlighting differences in spending patterns among group members. Meanwhile, the "Tip Ratio" column sheds light on tipping behavior, demonstrating how generous customers are about their overall bill.This degree of detail allows for the investigation of group dynamics, the evaluation of service quality, and the discovery of potential trends or patterns that would not be apparent if only the initial total bill amount were considered.

Question 4 – Grouping Analysis

```python
df_ttnc.head()
```

```python
# Using the .groupby() function,calculate the average fare and age for male and female passengers, 
# categorized by their town of embarkation.

df_ttnc.groupby(['Embarked','Sex']).agg({'Fare': 'mean', 'Age': 'mean'})
```

Note: The table displays some fascinating statistics about the passengers who boarded the Titanic from various cities and whether they were male or female. For example, it shows us that, on average, people who boarded the Titanic in C (as Cherbourg) paid more for their tickets than those who boarded in other cities such as S (as Southampton) or Q (as Queenstown).

Question 5 - Advanced Pivot Table Analysis

```python
df_ttnc.head()
```

```python
# Creating Pivot table, calculating the mean and median age for each Pclass.

df_ttnc.pivot_table(index='Pclass', values='Age', aggfunc={'Age':['mean','median']})
```

```python
# Within Pivot table, Incorporate additional hierarchical grouping such as gender or survival status.

df_ttnc.pivot_table(index=['Pclass','Sex','Survived'], values='Age', aggfunc={'Age':['mean','median']})
```

Note: We chose to group the data by passenger class and gender to understand how age varied across different groups on the Titanic. By analyzing age differences within each class, between men and women, we can gain insights into how age dynamics might have influenced people's experiences during the Titanic tragedy.
